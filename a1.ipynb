```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
co2es=pd.read_csv("CO2 emissions by sector - Finland.csv",skiprows=4)
co2et=pd.read_csv("Total CO2 emissions - Finland.csv",skiprows=4)
```

```python
pf0,pa0=plt.subplots(2,1,figsize=(6.4,4.8*2))
pa0[0].scatter(co2et["Unnamed: 0"],co2et["CO2 emissions"])
pa0[0].plot(co2et["Unnamed: 0"],co2et["CO2 emissions"])

secs=["Electricity and heat producers","Other energy industries", "Industry", "Transport", "Residential", "Agriculture", "Commercial and public services", "Final consumption not elsewhere specified"]
for s in secs:
    pa0[1].scatter(co2es["Unnamed: 0"],co2es[s],label=s)
    pa0[1].plot(co2es["Unnamed: 0"],co2es[s])
#pa0[1].legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=8)
pa0[1].legend(loc='upper center', bbox_to_anchor=(0.5,-.11),fontsize=8)
for a in pa0:
    a.set_xlabel("Year")
    a.set_ylabel("CO$_2$-emission (Mt)")
pa0[0].set_ylim([0,1.05*max(co2et["CO2 emissions"])])
pf0.tight_layout()
#pf0.savefig("a1_co2e.pdf")
```
